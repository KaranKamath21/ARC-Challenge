```python
import numpy as np
import matplotlib.pyplot as plt
import json

# Load the data from the JSON file
with open('task.json') as f:
    data = json.load(f)
```

```python
# Define the color rotation function
def rotate_colors(grid):
    return (grid + 1) % 3

# Define the function to visualize a grid
def visualize_grid(grid):
    fig, ax = plt.subplots()
    ax.imshow(grid, cmap='tab10')
    plt.show()
```

```python
# Process the training data
for i, task in enumerate(data['train']):
    input_grid = np.array(task['input'])
    output_grid = np.array(task['output'])

    print(f"Training Task {i+1}:")

    # Get the input grid
    print("Input Grid:")
    visualize_grid(input_grid)

    # Get the output grid
    print("Expected Output Grid:")
    visualize_grid(output_grid)
```

```python
# Process the test data
for i, task in enumerate(data['test']):
    input_grid = np.array(task['input'])
    output_grid = np.array(task['output'])

    print(f"Test Task {i+1}:")

    # Get the input grid
    print("Input Grid:")
    visualize_grid(input_grid)

    # Get the output grid
    print("Expected Output Grid:")
    visualize_grid(output_grid)

    # Get the prediction
    predicted_output_grid = rotate_colors(input_grid)
    print("Predicted Output Grid:")
    visualize_grid(predicted_output_grid)
```
